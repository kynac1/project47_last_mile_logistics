```python
import pandas as pd
import json
import matplotlib 
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 2}

matplotlib.rc('font', **font)
```

```python
with open("current_results3.json", 'r') as f:
    data = json.load(f)
data = {**{i:k.strip("()").replace("'","").split(', ')+[0]+[v[0]] for i,(k,v) in enumerate(data.items())} , **{i:k.strip("()").replace("'","").split(', ')+[1]+[v[1]] for i,(k,v) in enumerate(data.items())}, **{i:k.strip("()").replace("'","").split(', ')+[2]+[v[2]] for i,(k,v) in enumerate(data.items())}}
data
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
df = df.rename(columns={0:'fss', 1:'lsm', 2:'tlim', 3:'run', 4:'obj'})
df
```
